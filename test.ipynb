```python
import os
import numpy as np

# your function unchanged
def max_cosine_similarity_per_row(embeddings):
    X = np.asarray(embeddings, dtype=np.float32)
    if X.ndim != 2:
        raise ValueError(f"Expected 2D array, got {X.shape}")
    N = X.shape[0]
    if N < 2:
        raise ValueError("Need at least 2 embeddings.")

    norms = np.linalg.norm(X, axis=1, keepdims=True).clip(min=1e-12)
    Xn = X / norms

    S = Xn @ Xn.T
    np.fill_diagonal(S, -np.inf)

    best_j = np.argmax(S, axis=1)
    best_sim = S[np.arange(N), best_j]
    best_pct = np.clip(best_sim, 0, 1) * 100.0
    return best_j, best_sim, best_pct
```

```python
from imghdr import what as is_image  # Detects image types based on file headers
from pathlib import Path
import cv2


BASE_DIR = "/home/kasimov/notebooks/SenatorRepresentativeImagePipeline/Senate_Images"
pipeline = PreprocessingPipeline()

# Traverse all image files under the base directory
for file_path in Path(BASE_DIR).rglob("*"):
    if not file_path.is_file():
        continue

    try:
        img = cv2.imread(str(file_path))
        if img is None:
            print(f"[WARN] Unreadable image, deleting: {file_path}")
            os.remove(file_path)
            continue

        num_faces = pipeline.detect_number_of_faces(img)

        if num_faces != 1:
            print(f"[INFO] Deleting {file_path}  {num_faces} faces detected")
            os.remove(file_path)
        else:
            print(f"[OK] {file_path}  1 face detected")

    except Exception as e:
        print(f"[ERROR] Failed to process {file_path}: {e}")

```

```python
from pathlib import Path
import numpy as np
from utils import *

# Define source and destination directories
SOURCE_DIR = Path("/home/kasimov/notebooks/SenatorRepresentativeImagePipeline/Senate_Images")
DEST_DIR = Path("/home/kasimov/notebooks/SenatorRepresentativeImagePipeline/Senate_Embeddings")

# Ensure destination directory exists
DEST_DIR.mkdir(parents=True, exist_ok=True)

# Traverse each person directory
for person_dir in SOURCE_DIR.iterdir():
    if person_dir.is_dir():
        # Create destination filename
        dest_filename = f"{person_dir.name}_Facenet_Embeddings.npy"
        dest_path = DEST_DIR / dest_filename

        # Skip if file already exists
        if dest_path.exists():
            print(f"Embeddings already exist for {person_dir.name}, skipping.")
            continue

        # Get all embeddings from that person's directory
        embeddings, paths = get_embeddings_from_directory(person_dir, max_faces=None)

        # Skip if no embeddings
        if embeddings is None or len(embeddings) == 0:
            print(f"No embeddings found for {person_dir.name}, skipping.")
            continue

        # Save embeddings as .npy
        np.save(dest_path, embeddings)
        print(f"Saved {len(embeddings)} embeddings to {dest_path}")
```
